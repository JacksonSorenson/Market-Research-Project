```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime
```

```python
# Not Seperated by Commas, Use \t to Make Columns Readable
df = pd.read_csv('/Users/jacksonsorenson/Documents/Pyhton Projects/Market Research Project/Original CSV/marketing_campaign.csv', sep='\t')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Check to see Null Values
print(df.isnull().sum())
```

```python
# Replacing Null Values in Income with Income Median
df['Income'].fillna(df['Income'].median(), inplace=True)
print(df.isnull().sum())
```

```python
# Looking for Duplicates
print(df.duplicated().sum())
```

```python
# Capitalizing all Words in Marital_Status and Education
df['Marital_Status'] = df['Marital_Status'].str.capitalize()
df['Education'] = df['Education'].str.capitalize()
```

```python
# Change to Date Time
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
print(df['Dt_Customer'])
```

```python
# Feature Engineering

# Churn With the Threshold of 74 Days With no Purchase
churn_threshold = 74
df['Churn'] = (df['Recency'] > churn_threshold).astype(int)

# Calculate Age and drop 'Year_Birth'
df['Age'] = 2024 - df['Year_Birth']
df.drop('Year_Birth', axis=1, inplace=True)

# Calculate Total Spend across product categories
spend_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Total_Spend'] = df[spend_columns].sum(axis=1)

# Calculate Total Children in Household
df['Total_Children'] = df['Kidhome'] + df['Teenhome']

# Calculate Customer Tenure (in months) from 'Dt_Customer'
today = datetime.today()
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
df['Customer_Tenure'] = (today - df['Dt_Customer']).dt.days / 30  # Convert days to months

# Calculate Average Monthly Spend
df['Avg_Monthly_Spend'] = df['Total_Spend'] / df['Customer_Tenure']

# Calculate Recency in Days (time since the last interaction with the customer)
df['Recency_Days'] = df['Recency']

# Age Group
df['Age_Group'] = pd.cut(df['Age'], bins=[18, 30, 40, 50, 60, 100], labels=['18-30', '31-40', '41-50', '51-60', '61+'])

# Revenue By Product Category
product_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['Revenue'] = df[product_columns].sum(axis=1)

# Product Category
df['Product_Category'] = df[['MntWines', 'MntFruits', 'MntMeatProducts', 
                             'MntFishProducts', 'MntSweetProducts', 
                             'MntGoldProds']].apply(lambda x: ', '.join(x.astype(str)), axis=1)


df.head()
```
